```python
# set a variable
random_age = 50
```

```python
# test the variable with conditions
if random_age < 15:
    print('Child')
elif random_age < 25:
    print('Youth')
elif random_age<65:
    print('Adult')
else:
    print('Senior')
```

```python
# create dataframe conditions and choices
conditions = [users_df['age'] < 15, users_df['age'] < 25, users_df['age'] < 65]
choices = ['Child', 'Youth', 'Adult']

# create a new column for the result
users_df['age_bracket'] = np.select(conditions, choices, default='Senior')
```

### Joining tables in Python

Syntax:

    pd.merge(left_table, right_table, how, on)
    How: inner, left, right, full
    on: The primary - foreign key pair that you're joining it on
    When the key pairs have different names, you use,
    left_on
    right_on

```python
# read the order items table
order_items_df = pd.read_csv('order_items.csv')
```

```python
order_items_df.head()
```

```python
# read the product table
products_df = pd.read_csv('products.csv')
```

```python
products_df.head()
```

```python
# to join users table and order items table
```

```python
users_to_join = users_df.loc[:,['id','age_bracket']]
```

```python
age_bracket_merged = pd.merge(order_items_df, users_to_join, how='left', left_on='user_id', right_on='id')
```

```python
products_to_join = products_df.loc[:,['id','category']]
```

```python
fully_merged_table = pd.merge(age_bracket_merged, products_to_join, how='left', left_on='product_id', right_on='id')
```

```python
fully_merged_table.head()
```

```python
fully_merged_table.groupby(['age_bracket','category']).count()
```

```python
fully_merged_table.groupby(['category','age_bracket']).count()
```
